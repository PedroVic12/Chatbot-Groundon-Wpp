```python

import speech_recognition as sr
from gtts import gTTS
import os
import playsound
import wikipedia
import webbrowser
import pyjokes
import pyaudio
import os

# Desabilitar mensagens de erro ALSA e Jack
os.environ['PYGAME_HIDE_SUPPORT_PROMPT'] = 'hide'

class Robo:
    def __init__(self):
        self.recognizer = sr.Recognizer()
        self.microphone = sr.Microphone()

    def get_audio(self):
        with self.microphone as source:
            try:
                self.recognizer.pause_threshold = 1
                self.recognizer.adjust_for_ambient_noise(source, 1)
                audio = self.recognizer.listen(source)
                said = self.recognizer.recognize_google(audio)
                print(said)
                return said.lower()
            except Exception as e:
                print("Não entendi: " + str(e))
                return ""


    def salvar_para_arquivo(self, texto, nome_arquivo):
        with open(nome_arquivo, 'w') as arquivo:
            arquivo.write(texto)

    def ler_de_arquivo(self, nome_arquivo):
        try:
            with open(nome_arquivo, 'r') as arquivo:
                texto = arquivo.read()
                return texto
        except FileNotFoundError:
            return "O arquivo não foi encontrado."

    def falar(self, texto):
        tts = gTTS(text=texto, lang='pt')
        fileName = 'voice.mp3'
        try:
            os.remove(fileName)
        except OSError:
            pass
        tts.save(fileName)
        playsound.playsound(fileName)

    def executar(self):
        while True:
            print('Estou ouvindo...')
            texto = self.get_audio()
            if 'olá' in texto:
                self.falar('Olá, como posso ajudar?')
            elif 'youtube' in texto:
                self.falar('Abrindo YouTube')
                url = 'https://www.youtube.com/'
                webbrowser.get().open(url)
            elif 'pesquisar' in texto:
                self.falar('Pesquisando na Wikipedia')
                query = texto.replace('pesquisar', '')
                result = wikipedia.summary(query, sentences=3)
                self.falar('De acordo com a Wikipedia')
                print(result)
                self.falar(result)

            elif 'piada' in texto:
                self.falar(pyjokes.get_joke())

            elif 'salvar' in texto:
                self.falar('O que você gostaria de salvar em um arquivo de texto?')
                texto_para_salvar = self.get_audio()
                if texto_para_salvar:
                    self.salvar_para_arquivo(texto_para_salvar, 'texto_salvo.txt')
                    self.falar('Texto salvo com sucesso.')
            elif 'ler arquivo' in texto:
                texto_lido = self.ler_de_arquivo('texto_salvo.txt')
                self.falar(texto_lido)

            elif 'sair' in texto:
                self.falar('Até mais, Mestre Pedro')
                break


if __name__ == "__main__":
    robo = Robo()
    robo.executar()
```

```python
import pandas as pd
import os

def gerarJson(df, tabela_num):
    output_folder = 'Cardapio Json'  # Nome da pasta de saída
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    json_data = df.to_json(orient='records', force_ascii=False)
    file_name = f'tabela{tabela_num}.json'  # Nome do arquivo
    file_path = os.path.join(output_folder, file_name)  # Caminho completo do arquivo

    with open(file_path, 'w', encoding='utf-8') as file:
        file.write(json_data)

    print(f'Arquivo {file_name} salvo com sucesso!')

def criar_tabelas_dinamicas(colunas, df, tabela_num=1):
    tabela_colunas = []
    tabela_valores = []

    tabelas = []

    for coluna in colunas:
        if 'Unnamed' in coluna:
            if tabela_colunas:
                tabela = pd.DataFrame({col: val for col, val in zip(tabela_colunas, tabela_valores)})
                tabela = tabela.dropna()
                tabelas.append(tabela)
                tabela_colunas = []
                tabela_valores = []
                tabela_num += 1
            continue

        tabela_colunas.append(coluna)
        tabela_valores.append(df[coluna])

    if tabela_colunas:
        tabela = pd.DataFrame({col: val for col, val in zip(tabela_colunas, tabela_valores)})
        tabela = tabela.dropna()
        tabelas.append(tabela)

    for tabela_df in tabelas:
        display(tabela_df)
        gerarJson(tabela_df, tabela_num)
        tabela_num += 1

    return tabelas

colunas = ['Açaí e Pitaya', '300ml', '500ml', 'Unnamed: 3', 'Tapioca', 'Preço', 'Unnamed: 6']

tabelas = criar_tabelas_dinamicas(colunas_dataframe, cardapio_df)

print('\nProcesso concluído.')
```

```python
for tabela_nome, tabela_df in tabelas.items():
    print(f"{tabela_nome}:")
    display(tabela_df)
    gerarJson(tabela_df)
    print()
```
